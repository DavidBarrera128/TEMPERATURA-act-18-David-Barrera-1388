## 1. Fórmulas de Conversión

Para realizar estas conversiones, utilizamos las siguientes relaciones matemáticas:

* **De Grados Celsius ($^{\circ}C$) a Grados Fahrenheit ($^{\circ}F$):**
    $$F = (C \times \frac{9}{5}) + 32$$
    *(O de forma simplificada: $F = (C \times 1.8) + 32$)*

* **De Grados Fahrenheit ($^{\circ}F$) a Grados Celsius ($^{\circ}C$):**
    $$C = (F - 32) \times \frac{5}{9}$$
    *(O de forma simplificada: $C = \frac{F - 32}{1.8}$)*

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo 1: Convertir 25°C a Fahrenheit
Este es un valor común para una tarde agradable.

1.  **Identificar la fórmula:** $F = (C \times 1.8) + 32$.
2.  **Sustituir el valor:** $F = (25 \times 1.8) + 32$.
3.  **Multiplicar:** $25 \times 1.8 = 45$.
4.  **Sumar la constante:** $45 + 32 = 77$.
* **Resultado:** **77°F**.

### Ejemplo 2: Convertir 98.6°F a Celsius
Esta es la temperatura promedio del cuerpo humano.

1.  **Identificar la fórmula:** $C = (F - 32) / 1.8$.
2.  **Sustituir el valor:** $C = (98.6 - 32) / 1.8$.
3.  **Restar la constante:** $98.6 - 32 = 66.6$.
4.  **Dividir:** $66.6 / 1.8 = 37$.
* **Resultado:** **37°C**.

## 3. Código en Python: Convertidor Interactivo

Aquí tienes el programa siguiendo todas tus especificaciones. He separado la lógica en funciones para que el código sea limpio y profesional.

```python
def convertir_centigrados_a_fahrenheit():
    try:
        celsius = float(input("\nIngrese los grados Celsius a convertir: "))
        fahrenheit = (celsius * 1.8) + 32
        print(f">>> {celsius}°C equivale a {fahrenheit:.2f}°F")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def convertir_fahrenheit_a_celsius():
    try:
        fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
        celsius = (fahrenheit - 32) / 1.8
        print(f">>> {fahrenheit}°F equivale a {celsius:.2f}°C")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a centígrados")
        print("3. Salir")
        
        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == '1':
            convertir_centigrados_a_fahrenheit()
        elif opcion == '2':
            convertir_fahrenheit_a_celsius()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
```

### Notas sobre el código:
* **`float()`**: Permite que el usuario ingrese números decimales (ej. 36.5).
* **`try/except`**: Lo agregué como un "plus" para que el programa no se detenga si el usuario escribe letras en lugar de números por accidente.
* **`{valor:.2f}`**: Se usa para redondear el resultado a 2 decimales y que se vea más ordenado.
* **`while True` con `break`**: Mantiene el menú activo infinitamente hasta que se decide salir con la opción 3.

````python
def convertir_centigrados_a_fahrenheit():
    try:
        celsius = float(input("\nIngrese los grados Celsius a convertir: "))
        fahrenheit = (celsius * 1.8) + 32
        print(f">>> {celsius}°C equivale a {fahrenheit:.2f}°F")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def convertir_fahrenheit_a_celsius():
    try:
        fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
        celsius = (fahrenheit - 32) / 1.8
        print(f">>> {fahrenheit}°F equivale a {celsius:.2f}°C")
    except ValueError:
        print("Error: Por favor, ingrese un número válido.")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir grados centígrados a Fahrenheit")
        print("2. Convertir grados Fahrenheit a centígrados")
        print("3. Salir")

        opcion = input("\nSeleccione una opción (1-3): ")

        if opcion == '1':
            convertir_centigrados_a_fahrenheit()
        elif opcion == '2':
            convertir_fahrenheit_a_celsius()
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            print("!David Barrera NL = 8")
            break
        else:
            print("Opción no válida. Intente de nuevo.")

# Ejecución del programa
if __name__ == "__main__":
    menu()
````
